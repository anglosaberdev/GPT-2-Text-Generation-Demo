# Import Libraries
Importing necessary libraries for NLP and model visualization:

- **transformers**: provides pre-trained NLP models and easy-to-use pipelines
- **torch**: used for tensor operations
- **bertviz**: for visualizing attention in Transformer models

Set a fixed random seed to make text generation reproducible.

```python
from transformers import pipeline, set_seed, GPT2Tokenizer, GPT2LMHeadModel
from torch import tensor, numel
from torch import tensor, arange
set_seed(42)
```

# Text Generation using GPT-2
- We create a text generation pipeline using the pre-trained **GPT-2** model.
- The pipeline takes a **prompt** as input and generates text continuations.
- `max_length` sets the maximum number of tokens in the generated text.
- `num_return_sequences` controls how many different text variations are returned.

```python
generator = pipeline('text-generation', model='gpt2')
generator("Hello, I'm GPT2 model and I", max_length=30, num_return_sequences=3)
```

# Checking Tokens in GPT-2 Vocabulary
- We initialize the GPT-2 tokenizer using the pre-trained model.
- The tokenizer converts words into token IDs that the model understands.
- Using `'angelo' in tokenizer.get_vocab()` we check if the word **"angelo"** exists in GPT-2's vocabulary.
- Returns **True** if it exists, **False** if it doesn't.

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
'Angelo Saber' in tokenizer.get_vocab()
```

# Tokenizing and Decoding with GPT-2 Tokenizer
- First, the tokenizer converts a sentence into **token IDs** using `encode()`.
- Then, `convert_ids_to_tokens()` translates these IDs back into **tokens**.
- This shows how GPT-2 internally represents text as subword tokens.

```python
tokenizer.convert_ids_to_tokens(tokenizer.encode('What AI and machine learning allows you to do is find the needle in the haystack'))
```

# Encoding a Sentence into Token IDs
- We use `tokenizer.encode()` to convert a sentence into a sequence of **token IDs**.
- These IDs are the numerical representation that GPT-2 understands internally.
- This is the step before feeding text into the model for generation or analysis.


```python
tokenizer.encode('What AI and machine learning allows you to do is find the needle in the haystack')
```

# Loading GPT-2 Model and Tokenizer
- `GPT2Tokenizer.from_pretrained('gpt2')` loads the pre-trained GPT-2 tokenizer to convert text ↔ token IDs.
- `GPT2LMHeadModel.from_pretrained('gpt2')` loads the pre-trained GPT-2 language model for text generation.
- After this, `model` can take token IDs as input and generate predictions for the next tokens.

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model    = GPT2LMHeadModel.from_pretrained('gpt2')
```

# Counting Total Parameters in GPT-2
- Iterate over all parameters in the model using `model.parameters()`.
- Use `numel(param)` to get the number of elements in each parameter tensor.
- Sum them to get the total number of parameters in the model.
- Print the total count in a readable format with commas.

```python
total_params = 0
for param in model.parameters():
    total_params += numel(param)
    
print(f'Number of params: {total_params:,}')
```
